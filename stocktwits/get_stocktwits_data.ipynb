```python
def load_csv(path):
    '''
    Loads a csv of the given path and returns as a dataframe
    
    '''
    data = pd.read_csv(path)
    
    return data
    
```

```python
def format_json(val):
    '''
    Takes in a json values and formats it for our sentiment analysis model
    
    DATE | TICKER | MESSAGE BODY
    
    Returns formatted dataframe
    
    '''
    # initialize dataframe
    formatted = pd.DataFrame(columns={"date", "ticker", "message"})
    
    # extract each date, ticker, and message
    try :
        for i in range(len(val["messages"])):
            date = val["messages"][i]["created_at"]
            message = val["messages"][i]["body"]
            ticker = val["messages"][i]["symbols"][0]["symbol"]

            formatted.loc[len(formatted.index)] = [ticker, message, date]
    except:
        
        print("Could not find messages for ", val)
        
        
    return formatted
```

```python
def pull_stock_data(ticker):
    '''
    Uses stocktwits API to extract relevant posting data about the given stock ticker
    '''
    
    # connect to the api and load json data
    response = requests.get("https://api.stocktwits.com/api/2/streams/symbol/" + ticker + ".json")
    data = json.loads(response.content.decode(response.encoding))
    
    # format json value to a dataframe
    formatted_df = format_json(data)
    
    return formatted_df
```

```python
def get_stocks_data(stocks):
    '''
    Retrieves and returns formatted stocktwits data for the given list of stocks
    
    '''
    
    stock_df = pd.DataFrame()
    
    for stock in stocks:
        temp = pull_stock_data(stock)
        stock_df = stock_df.append(temp)
        stock_df.sort_values(by="date", inplace=True)
        stock_df.reset_index(drop=True, inplace=True)
    
    return stock_df
        
```
